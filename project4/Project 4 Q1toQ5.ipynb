### Q1) Plot a heatmap of Pearson correlation matrix of dataset columns. 
Report which features have the highest absolute correlation with the target variable and what that implies.

### Bikes 

```python
import seaborn as sns
bike_corr = bikedf.corr(method ='pearson')
fig, ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(
    bike_corr,
    annot = True, 
    vmin=-1, vmax=1, center=0,
    fmt='.2g',
    cmap='inferno',
    linewidths=1,
    linecolor='black',
    square=True
)
plt.savefig('bikeheatmap.jpg')
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

From the heatmap, we can observe that the highest absolute correlation with the target variables are as folllowing. <br>
Note: The three taret variables are not counted when observing the highest absolte correlations because the relationship between the tree is obvious: <br> &nbsp;&nbsp;&nbsp;&nbsp; cnt = casual + registered <br>
Casual:<br>
- temp(Normalized temperature in Celsius), 0.54 <br>
- atemp(Normalized feeling temperature in Celsius), 0.54 <br>
- also the workingday also has a very high absolute correlation, -0.52 <br>

Registered:<br>
- temp(Normalized temperature in Celsius), 0.54 <br>
- atemp(Normalized feeling temperature in Celsius), 0.54 <br>

Cnt:<br>
- temp(Normalized temperature in Celsius), 0.63 <br>
- atemp(Normalized feeling temperature in Celsius), 0.63 <br>

### Q4) For bike sharing dataset, plot the count number per day for a few months. Can you identify any repeating patterns in every month?

```python
def monthlyplot(year, month):
    monthcnt = bikedf.loc[bikedf['yr'] == year].loc[bikedf['mnth'] == month].cnt
    fig, ax = plt.subplots(figsize=(8,8))
    ax.bar(list(range(len(monthcnt))), monthcnt, width=0.3, color='plum')
    ax.set_xticks(list(range(len(monthcnt))))
    ax.set_xticklabels(list(range(len(monthcnt))), rotation=45, rotation_mode="anchor", ha="right")
    plt.show()
```

```python
for year in [0,1]:
    for month in range(1, 10):
        monthlyplot(0, month)
```
