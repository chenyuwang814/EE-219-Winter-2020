```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
dir = 'C:/Users/UETale/Desktop/UCLA/quarter2/EE219/project4/'
bike = pd.read_csv(str(dir) + 'Bike-Sharing-Dataset/'+ 'day.csv', header=0)
video = pd.read_csv(str(dir) + 'online_video_dataset/' + 'transcoding_mesurment.tsv', sep='\t', header=0)
```

```python
bike
```

```python
def onehot_encode(df,dummies_fields,drop_fields):
    for each in dummies_fields:
        dummies = pd.get_dummies(df.loc[:, each], prefix=each)
        df = pd.concat([df, dummies], axis=1)
    df = df.drop(drop_fields, axis=1)
    return df
```

```python
bike_onehot = onehot_encode(bike,dummies_fields = ['season', 'yr', 'mnth', 'weekday'],
                            drop_fields = ['dteday','season', 'yr', 'mnth', 'weekday','instant'])
```

```python
bike_onehot
```

```python
video
```

```python
video_onehot = onehot_encode(video,dummies_fields = ['codec', 'o_codec'],
                            drop_fields = ['id','codec','o_codec'])
```

```python
video_onehot
```

```python
bike_std = pd.DataFrame(preprocessing.scale(bike_onehot),columns = bike_onehot.columns)
video_std = pd.DataFrame(preprocessing.scale(video_onehot),columns = video_onehot.columns)
```

### mutual_info_regression

```python
from sklearn.feature_selection import mutual_info_regression
bike_std_featureMat = bike_std.drop(['casual','registered','cnt'],axis = 1)
```

```python
# mutual info regression of bike (3 tragets)
bike_casual_mir = pd.DataFrame(mutual_info_regression(bike_std_featureMat.values,bike_std['casual']).reshape(1,-1),columns = bike_std_featureMat.columns )
bike_registered_mir = pd.DataFrame(mutual_info_regression(bike_std_featureMat.values,bike_std['registered']).reshape(1,-1),columns = bike_std_featureMat.columns )
bike_cnt_mir = pd.DataFrame(mutual_info_regression(bike_std_featureMat.values,bike_std['cnt']).reshape(1,-1),columns = bike_std_featureMat.columns )
```

```python
video_std_featureMat = video_std.drop(['utime'],axis = 1)
```

```python
# mutual info regression of video (target:utime)
video_utime_mir = pd.DataFrame(mutual_info_regression(video_std_featureMat.values,video_std['utime']).reshape(1,-1),columns = video_std_featureMat.columns )
```
